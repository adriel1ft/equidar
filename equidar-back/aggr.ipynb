```python
import pandas as pd
```

```python
df = pd.read_csv("data/IDEB_ANOS_FINAIS_PB.csv")
df = df[['SG_UF', 'CO_MUNICIPIO', 'NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE']]
df['ORIGEM'] = 'IDEB_ANOS_FINAIS_PB'
df.head()
```

```python
df2 = pd.read_csv("data/IDEB_ANOS_INICIAIS_PB.csv")
df2 = df2[['SG_UF', 'CO_MUNICIPIO', 'NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE']]
df2['ORIGEM'] = 'IDEB_ANOS_INICIAIS_PB'
df2.head()
```

```python
df3 = pd.read_csv("data/IDEB_ENSINO_MEDIO_PB.csv")
df3 = df3[['SG_UF', 'CO_MUNICIPIO', 'NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE']]
df3['ORIGEM'] = 'IDEB_ENSINO_MEDIO_PB'
df3.head()
```

```python
# concatenate all schools, remove duplicates by setting a new category 
# - fundamental, anos iniciais
# - fundamental, anos finais
# - fundamental, anos iniciais e finais
# - medio

all_schools = pd.concat([df, df2, df3])
all_schools["CATEGORY"] = all_schools.groupby("ID_ESCOLA")["ORIGEM"].transform(
    lambda x: " & ".join(sorted(x.str.replace("IDEB_", "").str.replace("_PB", "").str.replace("ENSINO_", "").str.replace("ANOS_", "Anos ")))
).drop_duplicates()
all_schools = all_schools.drop_duplicates(subset=["ID_ESCOLA"])
print(all_schools.shape)
all_schools.head()
```
