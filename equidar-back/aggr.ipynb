```python
import pandas as pd
```

```python
df = pd.read_csv("data/IDEB_ANOS_FINAIS_PB.csv")
df = df[['SG_UF', 'CO_MUNICIPIO', 'NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE', 'VL_NOTA_MEDIA_2021']]
df['ORIGEM'] = 'IDEB_ANOS_FINAIS_PB'
df.head()

# complete historical data for null values in VL_NOTA_MEDIA_2021 with the data from VL_NOTA_MEDIA_2019
df_historical = pd.read_csv("data/IDEB_ANOS_FINAIS_PB.csv")
df_historical = df_historical[['ID_ESCOLA', 'VL_NOTA_MEDIA_2019', 'VL_NOTA_MEDIA_2017']]
df_historical.head()

# for the rows with null values in VL_NOTA_MEDIA_2021, fill with the corresponding value from VL_NOTA_MEDIA_2019
df = df.merge(df_historical, on='ID_ESCOLA', how='left')
df['VL_NOTA_MEDIA_2021'] = df.apply(lambda row: row['VL_NOTA_MEDIA_2019'] if pd.isnull(row['VL_NOTA_MEDIA_2021']) else row['VL_NOTA_MEDIA_2021'], axis=1)
df['VL_NOTA_MEDIA_2021'] = df.apply(lambda row: row['VL_NOTA_MEDIA_2017'] if pd.isnull(row['VL_NOTA_MEDIA_2021']) else row['VL_NOTA_MEDIA_2021'], axis=1)
df = df.drop(columns=['VL_NOTA_MEDIA_2019', 'VL_NOTA_MEDIA_2017'])
df.head()
```

```python
df2 = pd.read_csv("data/IDEB_ANOS_INICIAIS_PB.csv")
df2 = df2[['SG_UF', 'CO_MUNICIPIO', 'NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE', 'VL_NOTA_MEDIA_2021']]
df2['ORIGEM'] = 'IDEB_ANOS_FINAIS_PB'
df2.head()

# complete historical data for null values in VL_NOTA_MEDIA_2021 with the data from VL_NOTA_MEDIA_2019
df_historical = pd.read_csv("data/IDEB_ANOS_FINAIS_PB.csv")
df_historical = df_historical[['ID_ESCOLA', 'VL_NOTA_MEDIA_2019', 'VL_NOTA_MEDIA_2017']]
df_historical.head()

# for the rows with null values in VL_NOTA_MEDIA_2021, fill with the corresponding value from VL_NOTA_MEDIA_2019
df2 = df2.merge(df_historical, on='ID_ESCOLA', how='left')
df2['VL_NOTA_MEDIA_2021'] = df2.apply(lambda row: row['VL_NOTA_MEDIA_2019'] if pd.isnull(row['VL_NOTA_MEDIA_2021']) else row['VL_NOTA_MEDIA_2021'], axis=1)
df2['VL_NOTA_MEDIA_2021'] = df2.apply(lambda row: row['VL_NOTA_MEDIA_2017'] if pd.isnull(row['VL_NOTA_MEDIA_2021']) else row['VL_NOTA_MEDIA_2021'], axis=1)
df2 = df2.drop(columns=['VL_NOTA_MEDIA_2019', 'VL_NOTA_MEDIA_2017'])
df2.head()
```

```python
df3 = pd.read_csv("data/IDEB_ENSINO_MEDIO_PB.csv")
df3 = df3[['SG_UF', 'CO_MUNICIPIO', 'NO_MUNICIPIO', 'ID_ESCOLA', 'NO_ESCOLA', 'REDE', 'VL_NOTA_MEDIA_2021']]
df3['ORIGEM'] = 'IDEB_ANOS_FINAIS_PB'
df3.head()

# complete historical data for null values in VL_NOTA_MEDIA_2021 with the data from VL_NOTA_MEDIA_2019
df_historical = pd.read_csv("data/IDEB_ANOS_FINAIS_PB.csv")
df_historical = df_historical[['ID_ESCOLA', 'VL_NOTA_MEDIA_2019', 'VL_NOTA_MEDIA_2017']]
df_historical.head()

# for the rows with null values in VL_NOTA_MEDIA_2021, fill with the corresponding value from VL_NOTA_MEDIA_2019
df3 = df3.merge(df_historical, on='ID_ESCOLA', how='left')
df3['VL_NOTA_MEDIA_2021'] = df3.apply(lambda row: row['VL_NOTA_MEDIA_2019'] if pd.isnull(row['VL_NOTA_MEDIA_2021']) else row['VL_NOTA_MEDIA_2021'], axis=1)
df3['VL_NOTA_MEDIA_2021'] = df3.apply(lambda row: row['VL_NOTA_MEDIA_2017'] if pd.isnull(row['VL_NOTA_MEDIA_2021']) else row['VL_NOTA_MEDIA_2021'], axis=1)
df3 = df3.drop(columns=['VL_NOTA_MEDIA_2019', 'VL_NOTA_MEDIA_2017'])
df3.head()
```

```python
# concatenate all schools, remove duplicates by setting a new category 
# - fundamental, anos iniciais
# - fundamental, anos finais
# - fundamental, anos iniciais e finais
# - medio

all_schools = pd.concat([df, df2, df3]).drop_duplicates()
all_schools.groupby("ID_ESCOLA").size().value_counts()
```

```python
# Primeiro, vamos criar a categoria sem o drop_duplicates na transform
all_schools["CATEGORY"] = all_schools.groupby("ID_ESCOLA")["ORIGEM"].transform(
    lambda x: " & ".join(sorted(x.str.replace("IDEB_", "").str.replace("_PB", "").str.replace("ENSINO_", "").str.replace("ANOS_", "Anos ")))
)

# Depois removemos as duplicatas
all_schools = all_schools.drop_duplicates(subset=["ID_ESCOLA"])

print(all_schools.shape)
all_schools.head()
```

```python
all_schools.rename(columns={"VL_NOTA_MEDIA_2021": "IDEB_SCORE"}, inplace=True)
```

```python
all_schools.to_csv("data/ALL_SCHOOLS_PB_WITH_SCORES.csv", index=False)
```
